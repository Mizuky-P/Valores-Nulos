```python
#Importar librerias necesarias
import pandas as pd
```

```python
#Carga desde un archivo .csv sin indice
detalle_pyp=pd.read_excel("Detalle precios y productos fabricados 2022.xlsx")
detalle_pyp
```

```python
#Identificar valores nulos por columna
valores_nulos_detalle_pyp=detalle_pyp.isnull().sum()
valores_nulos_detalle_pyp
```

```python
#Realizamos una copia del dataframe
copia_detalle=detalle_pyp.copy()
```

```python
#Cuarto método de sustitución de valores nulos
#Sustituir valores nulos por un string en  concreto
copia_detalle["CVE_DOC"] = copia_detalle["CVE_DOC"].fillna("F0000")
copia_detalle["NOMBRE_VENDEDOR"]=copia_detalle["NOMBRE_VENDEDOR"].fillna("VENDEDOR X")
copia_detalle["NOMBRE_CLIENTE"]=copia_detalle["NOMBRE_CLIENTE"].fillna("CLIENTE X")
copia_detalle["DESCR"]=copia_detalle["DESCR"].fillna("SIN DESCRIPCION")
copia_detalle["CVE_ART"]=copia_detalle["CVE_ART"].fillna("F000")
copia_detalle["FECHA_DOC"]=copia_detalle["FECHA_DOC"].fillna("XXXX-XX-XX")
```

```python
#Corroboramos valores nulos
valores_nulos=copia_detalle.isnull().sum()
valores_nulos
```

```python
#Primer método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media
copia_detalle["CANT"]=copia_detalle["CANT"].fillna(round(copia_detalle["CANT"].mean(),1))
copia_detalle["PRECIO_UNITARIO"]=copia_detalle["PRECIO_UNITARIO"].fillna(round(copia_detalle["PRECIO_UNITARIO"].mean(),1))
copia_detalle["SUBTOTAL_PARTIDA"]=copia_detalle["SUBTOTAL_PARTIDA"].fillna(round(copia_detalle["SUBTOTAL_PARTIDA"].mean(),1))
copia_detalle["COSTO_UNITARIO"]=copia_detalle["COSTO_UNITARIO"].fillna(round(copia_detalle["COSTO_UNITARIO"].mean(),1))
copia_detalle["COSTO_UNITARIO_CALCULADO"]=copia_detalle["COSTO_UNITARIO_CALCULADO"].fillna(round(copia_detalle["COSTO_UNITARIO_CALCULADO"].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=copia_detalle.isnull().sum()
valores_nulos
```

```python
#Convertir archivo filtrado a CSV
copia_detalle.to_csv("Detalle_precios_yp_roductos.csv")
```
