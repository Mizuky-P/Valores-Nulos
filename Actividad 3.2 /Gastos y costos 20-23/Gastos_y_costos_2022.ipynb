```python
#Importar librerias necesarias
import pandas as pd
```

```python
#Carga desde un archivo .csv sin indice y eliminar filas innecesarias
gastos_costos_22=pd.read_excel("Gastos y costos 20-23.xlsx",sheet_name="2022", skiprows=3)
gastos_costos_22
```

```python
#Identificar valores nulos por columna
valores_nulos_gastos_costos_22=gastos_costos_22.isnull().sum()
valores_nulos_gastos_costos_22
```

```python
#Realizamos una copia del dataframe
copia_gastos_costos_22= gastos_costos_22.copy()
```

```python
#Primer método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media
copia_gastos_costos_22["TC"]=copia_gastos_costos_22["TC"].fillna(round(copia_gastos_costos_22["TC"].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=copia_gastos_costos_22.isnull().sum()
valores_nulos
```

```python
#Tercer método de sustitución de valores nulos
#Sustituir valores nulos por un valor numérico en  concreto
copia_gastos_costos_22["Folio"] = copia_gastos_costos_22["Folio"].fillna("F0000")
copia_gastos_costos_22["Otros"] = copia_gastos_costos_22["Otros"].fillna("N/A")
copia_gastos_costos_22["Poliza"] = copia_gastos_costos_22["Poliza"].fillna("N/A")
```

```python
#Corroboramos valores nulos
valores_nulos=copia_gastos_costos_22.isnull().sum()
valores_nulos
```

```python
#Sexto método de sustitución de valores nulos
#Sustituir valores nulos por valores no nulos hacia adelante con "forward fill" ("ffill")
copia_gastos_costos_22["MP"] = copia_gastos_costos_22["MP"].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos=copia_gastos_costos_22.isnull().sum()
valores_nulos
```

```python
#Convertir archivo filtrado a CSV
copia_gastos_costos_22.to_csv("gastos_costos_22.csv")
```
