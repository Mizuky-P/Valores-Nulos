```python
#Importar librerias necesarias
import pandas as pd
```

```python
#Carga desde un archivo .csv sin indice y eliminar filas innecesarias
gastos_costos_23=pd.read_excel("Gastos y costos 20-23.xlsx",sheet_name="2023",skiprows=3)
gastos_costos_23
```

```python
#Identificar valores nulos por columna
valores_nulos_gastos_costos_23=gastos_costos_23.isnull().sum()
valores_nulos_gastos_costos_23
```

```python
#Realizamos una copia del dataframe
copia_gastos_costos_23 = gastos_costos_23.copy()
```

```python
#Cuarto método de sustitución de valores nulos
#Sustituir valores nulos por un string en  concreto
copia_gastos_costos_23["Folio"]=copia_gastos_costos_23["Folio"].fillna("F0000")
copia_gastos_costos_23["TIPO GASTO"]=copia_gastos_costos_23["TIPO GASTO"].fillna("INFO FALTANTE")
copia_gastos_costos_23["Poliza"]=copia_gastos_costos_23["Poliza"].fillna("Sin Poliza")
copia_gastos_costos_23["MP"]=copia_gastos_costos_23["MP"].fillna("SIN REGISTRO")
```

```python
#Corroboramos valores nulos
valores_nulos=copia_gastos_costos_23.isnull().sum()
valores_nulos
```

```python
#Tercer método de sustitución de valores nulos
#Sustituir valores nulos por un valor numérico en  concreto
copia_gastos_costos_23["Otros"]=copia_gastos_costos_23["Otros"].fillna("0")
```

```python
#Corroboramos valores nulos
valores_nulos=copia_gastos_costos_23.isnull().sum()
valores_nulos
```

```python
#Primer método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media
copia_gastos_costos_23["FP"]=copia_gastos_costos_23["FP"].fillna(round(copia_gastos_costos_23['FP'].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=copia_gastos_costos_23.isnull().sum()
valores_nulos
```

```python
#Convertir archivo filtrado a CSV
copia_gastos_costos_23.to_csv("gastos_costos_23.csv")
```
