```python
#Importar librerias necesarias
import pandas as pd
```

```python
#Carga desde un archivo .csv sin indice y eliminar filas innecesarias
gastos_costos_20=pd.read_excel("Gastos y costos 20-23.xlsx",sheet_name="2020", skiprows=5)
gastos_costos_20
```

```python
#Identificar valores nulos por columna
valores_nulos_gastos_costos_20=gastos_costos_20.isnull().sum()
valores_nulos_gastos_costos_20
```

```python
#Realizamos una copia del dataframe
copia_gastos_costos_20= gastos_costos_20.copy()
```

```python
#Cuarto método de sustitución de valores nulos
#Sustituir valores nulos por un string en  concreto
copia_gastos_costos_20["FOLIO"]=copia_gastos_costos_20["FOLIO"].fillna("F0000")
copia_gastos_costos_20["GASTO"]=copia_gastos_costos_20["GASTO"].fillna("INDEFINIDO")
copia_gastos_costos_20["POLIZA"]=copia_gastos_costos_20["POLIZA"].fillna("SIN POLIZA")
copia_gastos_costos_20["TIPO"]=copia_gastos_costos_20["TIPO"].fillna("INFO FALTANTE")
```

```python
#Corroboramos valores nulos
valores_nulos=copia_gastos_costos_20.isnull().sum()
valores_nulos
```

```python
#Quinto método de sustitución de valores nulos
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
copia_gastos_costos_20["TC"]=copia_gastos_costos_20["TC"].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos=copia_gastos_costos_20.isnull().sum()
valores_nulos
```

```python
#Primer método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media
copia_gastos_costos_20["IMPORTE"]=copia_gastos_costos_20["IMPORTE"].fillna(round(copia_gastos_costos_20['IMPORTE'].mean(),1))
copia_gastos_costos_20["IVA"]=copia_gastos_costos_20["IVA"].fillna(round(copia_gastos_costos_20['IVA'].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=copia_gastos_costos_20.isnull().sum()
valores_nulos
```

```python
#Convertir archivo filtrado a CSV
copia_gastos_costos_20.to_csv("gastos_costos_20.csv")
```
