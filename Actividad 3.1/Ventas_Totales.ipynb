```python
#Importar librerias necesarias
import pandas as pd
```

```python
#Carga desde un archivo .csv sin indice
ventas_totales = pd.read_csv("Ventas_totales.csv")
ventas_totales
```

```python
# Identificar y contar valores nulos por columna
valores_nulos_ventas_totales = ventas_totales.isnull().sum()
valores_nulos_ventas_totales
```

```python
#Realizamos una copia del dataframe
copia_ventas_totales=ventas_totales.copy()
```

```python
#Tercer método de sustitución de valores nulos
#Sustituir valores nulos por un valor numérico en  concreto
copia_ventas_totales["salon_ventas"]= copia_ventas_totales["salon_ventas"].fillna(0)
copia_ventas_totales["tarjetas_credito"]= copia_ventas_totales["tarjetas_credito"].fillna(0)
copia_ventas_totales["otros"]= copia_ventas_totales["otros"].fillna(0)
copia_ventas_totales["almacen"]= copia_ventas_totales["almacen"].fillna(0)
copia_ventas_totales["indumentaria_calzado_textiles_hogar"]= copia_ventas_totales["indumentaria_calzado_textiles_hogar"].fillna(0)
copia_ventas_totales["electronicos_articulos_hogar"]= copia_ventas_totales["electronicos_articulos_hogar"].fillna(0)
```

```python
#Corroboramos valores nulos
valores_nulos=copia_ventas_totales.isnull().sum()
valores_nulos
```

```python
#Quinto método de sustitución de valores nulos
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
copia_ventas_totales["tarjetas_debito"] =copia_ventas_totales["tarjetas_debito"].fillna(method="ffill")
copia_ventas_totales["otros_medios"] =copia_ventas_totales["otros_medios"].fillna(method="ffill")
copia_ventas_totales["bebidas"] =copia_ventas_totales["bebidas"].fillna(method="ffill")
copia_ventas_totales["lacteos"] =copia_ventas_totales["lacteos"].fillna(method="ffill")
copia_ventas_totales["panaderia"] =copia_ventas_totales["panaderia"].fillna(method="ffill")
copia_ventas_totales["carnes"] =copia_ventas_totales["carnes"].fillna(method="ffill")
copia_ventas_totales["verduleria_fruteria"] =copia_ventas_totales["verduleria_fruteria"].fillna(method="ffill")
copia_ventas_totales["alimentos_preparados_rotiseria"] =copia_ventas_totales["alimentos_preparados_rotiseria"].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos=copia_ventas_totales.isnull().sum()
valores_nulos
```

```python
#Segundo método de sustitución de valores nulos
#Sustituir valores nulos con mediana
copia_ventas_totales["subtotal_ventas_alimentos_bebidas"]=copia_ventas_totales["subtotal_ventas_alimentos_bebidas"].fillna(round(copia_ventas_totales["subtotal_ventas_alimentos_bebidas"].median(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=copia_ventas_totales.isnull().sum()
valores_nulos
```

```python
#Convertir archivo filtrado a CSV
copia_ventas_totales.to_csv("Ventas Totales.csv")
```
